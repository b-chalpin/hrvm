```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# config
HRV_DANGER_THRESHOLD = 30
NUM_SAMPLES = 300

# generate random HRVs
hrv_samples = np.random.uniform(low=10, high=100, size=(NUM_SAMPLES,)).reshape(-1, 1)

def determine_hrv_dangerous(hrv):
    if hrv <= HRV_DANGER_THRESHOLD:
        return 1
    else:
        return 0

# generate labels for random data
hrv_labels = np.array(list(map(determine_hrv_dangerous, hrv_samples)))

# get training/test split
X_train, X_test, y_train, y_test = train_test_split(hrv_samples, hrv_labels, test_size=0.5, random_state=420)

lr = LogisticRegression()

# fit the model with data
lr.fit(X_train, y_train)

y_hat = lr.predict(X_test)

x_axis = np.arange(len(y_hat))

col = list(map(lambda x: "k" if x == 0 else "r", y_hat))

plt.scatter(x=x_axis, y=X_test, c=col)
plt.show()
```
